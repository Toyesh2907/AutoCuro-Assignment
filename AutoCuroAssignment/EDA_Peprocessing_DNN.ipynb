```python
import numpy as np
import pandas as pd
dataset=pd.read_csv('C:/Users/Lenovo/AutoCuroAssignment/tic_tac_toe_records_minimax.csv')
```

```python
dataset['Input']=None
for index, row in dataset.iterrows():
    str=dataset.iloc[index][0]
    input_data=[]
    for i in range(len(str)):
        if(str[i]=='         '):
            dataset.drop(index,inplace=True)
        if(str[i]=='X'):
            input_data.append(1)
        elif(str[i]=='O'):
            input_data.append(2)
        else:
            input_data.append(0)
    dataset.at[index,'Input']=input_data
```
